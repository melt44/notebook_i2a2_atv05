Implementando funÃ§Ãµes para atender Ã s seguintes perguntas:
- Quantos ninhos hÃ¡ no total?
- Qual a mÃ©dia de ovos por ninho com risco ðŸŸ¢?
- Quantos ninhos estÃ£o prestes a eclodir (dias â‰¤ 5)?

```python
#funÃ§Ã£o para contar a quantidade de ninhos registrados.

def contar_ninhos():
    global quantidade_ninhos
    quantidade_ninhos = len(ninhos_registrados)
    return print(f"Quantidade total de ninhos: {quantidade_ninhos}")

# Chamando a funÃ§Ã£o
contar_ninhos()
```

```python
#funÃ§Ã£o para calcular a mÃ©dia de ovos por ninho com risco ðŸŸ¢

def media_ovos_risco():
    global media_ovos
    total_ovos = 0
    quantidade_ninhos_risco_verde = 0

    for n in ninhos_registrados:
        if n['risco'] == "ðŸŸ¢":
            total_ovos += int(n['quantidade_ovos'])
            quantidade_ninhos_risco_verde += 1
    if quantidade_ninhos_risco_verde > 0:
        media_ovos = total_ovos / quantidade_ninhos_risco_verde
        return print(f"MÃ©dia de ovos por ninho com risco ðŸŸ¢: {media_ovos:.2f}")
    else:
        return print("Nenhum ninho com risco ðŸŸ¢ encontrado.")
    
# Chamando a funÃ§Ã£o
media_ovos_risco()
```

```python
#funÃ§Ã£o para contar a quantidade de ninhos prestes a eclodir

def ninhos_prestes_a_eclodir():
    global quantidade_ninhos_prestes_a_eclodir
    quantidade_ninhos_prestes_a_eclodir = 0

    for n in ninhos_registrados:
        if n['dias_para_eclosao'] <= 5:
            quantidade_ninhos_prestes_a_eclodir += 1
    return print(f"Quantidade de ninhos prestes a eclodir: {quantidade_ninhos_prestes_a_eclodir}")

# Chamando a funÃ§Ã£o
ninhos_prestes_a_eclodir()
```

```python
#funcao para encontrar a regiÃ£o com mais ninhos sob risco ðŸ”´

#eu poderia simplificar o cÃ³dido utilizando Counter, biblioteca padrÃ£o do python, para contar as ocorrÃªncias.
#from collections import Counter 

def regiao_mais_ninhos_risco_vermelho():
    global lista_regioes
    lista_regioes = []
    
    for n in ninhos_registrados:
        if n['risco'] == "ðŸ”´":
            lista_regioes.append(n['regiao'])
    
    if not lista_regioes:
        return (None, 0)

    contagens = {}
    for regiao in lista_regioes:
        contagens[regiao] = contagens.get(regiao, 0) + 1

    regiao_vencedora = None
    contagem_maxima = -1  

    for regiao, contagem in contagens.items():
        if contagem > contagem_maxima:
            contagem_maxima = contagem
            regiao_vencedora = regiao

    print(f"A regiÃ£o com mais ninhos sob risco ðŸ”´ Ã©: {regiao_vencedora} com {contagem_maxima} ocorrÃªncias.")

# Chamando a funÃ§Ã£o
regiao_mais_ninhos_risco_vermelho()
```

```python
#funcao para contar a quantidade de ninhos danificados e com predadores

def quantidade_ninhos_danificados():

    numero_de_ninhos_danificados = 0

    for n in ninhos_registrados:
        if n['predadores'] == True and n['status'] == 'danificado':
            numero_de_ninhos_danificados += 1
            print(n)
    print(f"NÃºmero de ninhos danificados e com predadores: {numero_de_ninhos_danificados}")

# Chamando a funÃ§Ã£o
quantidade_ninhos_danificados()
```

```python
#essa funÃ§Ã£o percorre a lista de ninhos registrados e exibe as informaÃ§Ãµes de cada ninho.

def relatorio_semana():
    print("Ninhos registrados:")
    for n in ninhos_registrados:
        print(f"RegiÃ£o: {n['regiao']}, Ovos: {n['quantidade_ovos']}, Status: {n['status']}, Risco: {n['risco']}, Dias para eclosÃ£o: {n['dias_para_eclosao']}, Predadores: {'Sim' if n['predadores'] else 'NÃ£o'}")

#funÃ§Ã£o para exibir o menu interativo 

def menu_interativo():
    while True:
        print("\nMenu interativo: ")
        print("1. Inserir novo ninho. ")
        print("2. Vizualizar o relatÃ³rio completo da semana. ")
        print("3. Consultar estatÃ­sticas. ")
        print("\n Digite 'Sair' para finalizar o menu. \n")

        opcao = input("Escolha uma opÃ§Ã£o: ").lower().strip()
        if opcao == "1":
            
            print("Inserindo novo ninho...")

            # Tratamento de dados individual e integrado a seÃ§Ã£o de inserÃ§Ã£o de novo ninho 

            # ValidaÃ§Ã£o da //RegiÃ£o
            while True:
                regiao = input("Digite a regiÃ£o do ninho (ex: 'Praia Norte'): ").strip()
                if regiao:
                    break
                print("A regiÃ£o nÃ£o pode estar em branco. Tente novamente.")

            # ValidaÃ§Ã£o da //Quantidade de Ovos
            while True:
                try:
                    quantidade_ovos = int(input("Digite a quantidade de ovos: "))
                    if quantidade_ovos >= 0:
                        break
                    print("A quantidade de ovos deve ser um nÃºmero positivo.")
                except ValueError:
                    print("Entrada invÃ¡lida. Por favor, digite um nÃºmero inteiro.")

            # ValidaÃ§Ã£o do //Status
            while True:
                status = input("Digite o status do ninho (intacto/danificado): ").lower().strip()
                if status in ["intacto", "danificado"]:
                    break
                print("Status invÃ¡lido. Deve ser 'intacto' ou 'danificado'.")

            # ValidaÃ§Ã£o do //Risco
            while True:
                relacao_risco = {"1": "ðŸŸ¢", "2": "ðŸŸ¡", "3": "ðŸ”´"}
                input_risco = input("Digite o risco do ninho (1 - ðŸŸ¢ / 2 - ðŸŸ¡ / 3 - ðŸ”´): ").strip()
                if input_risco in relacao_risco:
                    risco = relacao_risco[input_risco]
                    break
                print("Risco invÃ¡lido. Escolha 1, 2 ou 3.")

            # ValidaÃ§Ã£o dos //Dias para EclosÃ£o
            while True:
                try:
                    dias_para_eclosao = int(input("Digite os dias para eclosÃ£o: "))
                    if dias_para_eclosao >= 0:
                        break
                    print("Os dias para eclosÃ£o devem ser um nÃºmero positivo.")
                except ValueError:
                    print("Entrada invÃ¡lida. Por favor, digite um nÃºmero inteiro.")

            # ValidaÃ§Ã£o de //Predadores
            while True:
                predadores_input = input("O ninho tem predadores? (sim/nao): ").lower().strip()
                if predadores_input in ["sim", "s"]:
                    predadores = True
                    break
                elif predadores_input in ["nao", "n"]:
                    predadores = False
                    break
                print("Resposta invÃ¡lida. Por favor, digite 'sim' ou 'nao'.")

            # Adicionando os dados coletados na lista de ninhos 
            novo_ninho = {
                "regiao": regiao,
                "quantidade_ovos": quantidade_ovos,
                "status": status,
                "risco": risco,
                "dias_para_eclosao": dias_para_eclosao,
                 "predadores": predadores
                }
            ninhos_registrados.append(novo_ninho)
            print("Novo ninho adicionado com sucesso!")

        elif opcao == "2":
            print("RelatÃ³rio completo da semana:\n")
            relatorio_semana()
        
        elif opcao == "3":

            print("Mostrando estatÃ­sticas...\n")
            print("Quantidade total de ninhos:")
            contar_ninhos()
            media_ovos_risco()
            ninhos_prestes_a_eclodir()
            regiao_mais_ninhos_risco_vermelho()
            quantidade_ninhos_danificados()

        
        elif opcao in ["sair"]:

            print("Saindo do menu...")
            break

# Chamando a funÃ§Ã£o
menu_interativo()
```

Implementando fun√ß√µes para atender √†s seguintes perguntas:
- Quantos ninhos h√° no total?
- Qual a m√©dia de ovos por ninho com risco üü¢?
- Quantos ninhos est√£o prestes a eclodir (dias ‚â§ 5)?
- Qual regi√£o tem mais ninhos sob risco üî¥?
- Quantos ninhos t√™m presen√ßa de predadores e est√£o danificados?

```python
#fun√ß√£o para contar a quantidade de ninhos registrados.

def contar_ninhos():
    global quantidade_ninhos
    quantidade_ninhos = len(ninhos_registrados)
    return print(f"Quantidade total de ninhos: {quantidade_ninhos}")

# Chamando a fun√ß√£o
contar_ninhos()
```

```python
#fun√ß√£o para calcular a m√©dia de ovos por ninho com risco üü¢

def media_ovos_risco():
    global media_ovos
    total_ovos = 0
    quantidade_ninhos_risco_verde = 0

    for n in ninhos_registrados:
        if n['risco'] == "üü¢":
            total_ovos += int(n['quantidade_ovos'])
            quantidade_ninhos_risco_verde += 1
    if quantidade_ninhos_risco_verde > 0:
        media_ovos = total_ovos / quantidade_ninhos_risco_verde
        return print(f"M√©dia de ovos por ninho com risco üü¢: {media_ovos:.2f}")
    else:
        return print("Nenhum ninho com risco üü¢ encontrado.")
    
# Chamando a fun√ß√£o
media_ovos_risco()
```

```python
#fun√ß√£o para contar a quantidade de ninhos prestes a eclodir

def ninhos_prestes_a_eclodir():
    global quantidade_ninhos_prestes_a_eclodir
    quantidade_ninhos_prestes_a_eclodir = 0

    for n in ninhos_registrados:
        if n['dias_para_eclosao'] <= 5:
            quantidade_ninhos_prestes_a_eclodir += 1
    return print(f"Quantidade de ninhos prestes a eclodir: {quantidade_ninhos_prestes_a_eclodir}")

# Chamando a fun√ß√£o
ninhos_prestes_a_eclodir()
```

```python
#funcao para encontrar a regi√£o com mais ninhos sob risco üî¥

#eu poderia simplificar o c√≥dido utilizando Counter, biblioteca padr√£o do python, para contar as ocorr√™ncias.
#from collections import Counter 

def regiao_mais_ninhos_risco_vermelho():
    global lista_regioes
    lista_regioes = []
    
    for n in ninhos_registrados:
        if n['risco'] == "üî¥":
            lista_regioes.append(n['regiao'])
    
    if not lista_regioes:
        return (None, 0)

    contagens = {}
    for regiao in lista_regioes:
        contagens[regiao] = contagens.get(regiao, 0) + 1

    regiao_vencedora = None
    contagem_maxima = -1  

    for regiao, contagem in contagens.items():
        if contagem > contagem_maxima:
            contagem_maxima = contagem
            regiao_vencedora = regiao

    print(f"A regi√£o com mais ninhos sob risco üî¥ √©: {regiao_vencedora} com {contagem_maxima} ocorr√™ncias.")

# Chamando a fun√ß√£o
regiao_mais_ninhos_risco_vermelho()
```

```python
#funcao para contar a quantidade de ninhos danificados e com predadores

def quantidade_ninhos_danificados():

    numero_de_ninhos_danificados = 0

    for n in ninhos_registrados:
        if n['predadores'] == True and n['status'] == 'danificado':
            numero_de_ninhos_danificados += 1
            print(n)
    print(f"N√∫mero de ninhos danificados e com predadores: {numero_de_ninhos_danificados}")

# Chamando a fun√ß√£o
quantidade_ninhos_danificados()
```

```python
#essa fun√ß√£o percorre a lista de ninhos registrados e exibe as informa√ß√µes de cada ninho.

def relatorio_semana():
    print("Ninhos registrados:")
    for n in ninhos_registrados:
        print(f"Regi√£o: {n['regiao']}, Ovos: {n['quantidade_ovos']}, Status: {n['status']}, Risco: {n['risco']}, Dias para eclos√£o: {n['dias_para_eclosao']}, Predadores: {'Sim' if n['predadores'] else 'N√£o'}")

#fun√ß√£o para exibir o menu interativo 

def menu_interativo():
    while True:
        print("\nMenu interativo: ")
        print("1. Inserir novo ninho. ")
        print("2. Vizualizar o relat√≥rio completo da semana. ")
        print("3. Consultar estat√≠sticas. ")
        print("\n Digite 'Sair' para finalizar o menu. \n")

        opcao = input("Escolha uma op√ß√£o: ").lower().strip()
        if opcao == "1":
            
            print("Inserindo novo ninho...")

            # Tratamento de dados individual e integrado a se√ß√£o de inser√ß√£o de novo ninho 

            # Valida√ß√£o da //Regi√£o
            while True:
                regiao = input("Digite a regi√£o do ninho (ex: 'Praia Norte'): ").strip()
                if regiao:
                    break
                print("A regi√£o n√£o pode estar em branco. Tente novamente.")

            # Valida√ß√£o da //Quantidade de Ovos
            while True:
                try:
                    quantidade_ovos = int(input("Digite a quantidade de ovos: "))
                    if quantidade_ovos >= 0:
                        break
                    print("A quantidade de ovos deve ser um n√∫mero positivo.")
                except ValueError:
                    print("Entrada inv√°lida. Por favor, digite um n√∫mero inteiro.")

            # Valida√ß√£o do //Status
            while True:
                status = input("Digite o status do ninho (intacto/danificado): ").lower().strip()
                if status in ["intacto", "danificado"]:
                    break
                print("Status inv√°lido. Deve ser 'intacto' ou 'danificado'.")

            # Valida√ß√£o do //Risco
            while True:
                relacao_risco = {"1": "üü¢", "2": "üü°", "3": "üî¥"}
                input_risco = input("Digite o risco do ninho (1 - üü¢ / 2 - üü° / 3 - üî¥): ").strip()
                if input_risco in relacao_risco:
                    risco = relacao_risco[input_risco]
                    break
                print("Risco inv√°lido. Escolha 1, 2 ou 3.")

            # Valida√ß√£o dos //Dias para Eclos√£o
            while True:
                try:
                    dias_para_eclosao = int(input("Digite os dias para eclos√£o: "))
                    if dias_para_eclosao >= 0:
                        break
                    print("Os dias para eclos√£o devem ser um n√∫mero positivo.")
                except ValueError:
                    print("Entrada inv√°lida. Por favor, digite um n√∫mero inteiro.")

            # Valida√ß√£o de //Predadores
            while True:
                predadores_input = input("O ninho tem predadores? (sim/nao): ").lower().strip()
                if predadores_input in ["sim", "s"]:
                    predadores = True
                    break
                elif predadores_input in ["nao", "n"]:
                    predadores = False
                    break
                print("Resposta inv√°lida. Por favor, digite 'sim' ou 'nao'.")

            # Adicionando os dados coletados na lista de ninhos 
            novo_ninho = {
                "regiao": regiao,
                "quantidade_ovos": quantidade_ovos,
                "status": status,
                "risco": risco,
                "dias_para_eclosao": dias_para_eclosao,
                 "predadores": predadores
                }
            ninhos_registrados.append(novo_ninho)
            print("Novo ninho adicionado com sucesso!")

        elif opcao == "2":
            print("Relat√≥rio completo da semana:\n")
            relatorio_semana()
        
        elif opcao == "3":

            print("Mostrando estat√≠sticas...\n")
            print("Quantidade total de ninhos:")
            contar_ninhos()
            media_ovos_risco()
            ninhos_prestes_a_eclodir()
            regiao_mais_ninhos_risco_vermelho()
            quantidade_ninhos_danificados()

        
        elif opcao in ["sair"]:

            print("Saindo do menu...")
            break

# Chamando a fun√ß√£o
menu_interativo()
```
